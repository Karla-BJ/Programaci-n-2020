# **Primer** **punto**

```
#Preguntas código y textos#
saludo= ("Hola!, bienvenido a mi exámen final")
cantidadpesos= ("Porfavor ingrese la cantidad de pesos colombianos que desea convertir en dólares : ")
entradaNoValida= ("Los datos ingresados son inválidos, recuerda ingresar una cantidad numérica de pesos colombianos")
despedida= ("Gracias por hacer uso de nuestros servicios ♥")


print (saludo)

pesos = None
while (pesos == None):
  try:
    pesos= float(input(cantidadpesos))
  except ValueError:
    print (entradaNoValida)

tasaCambio= 0.000263157895

conversion= pesos*tasaCambio

print (conversion)

print (despedida)

```

# **Tercer** **punto**




```

import matplotlib.pyplot as plt

Meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
Ingresos = [10000000, 10000000, 8000000, 7000000, 6000000, 5000000, 4000000, 4000000, 3000000, 2000000, 2000000, 1000000]
plt.bar (Meses, Ingresos)
plt.title('Ingresos de Dúlce María durante el año 2020')
plt.xlabel('Meses')
plt.ylabel('Ingresos mensuales')
plt.show()
```

# **Cuarto** **punto**

```
import matplotlib.pyplot as plt
ciudades = ["Bogotá", "Medellín", "Cali", "Barranquilla", "Cartagena"]
kmCuad= [1775, 382, 619, 154, 609]
explodeList= [0.1,0,0,0,0]
plt.title ("Superficie en km² de las ciudades principales de Colombia ")
plt.pie (kmCuad, labels= ciudades, explode= explodeList, shadow= True, startangle=45)
plt.show()
```
