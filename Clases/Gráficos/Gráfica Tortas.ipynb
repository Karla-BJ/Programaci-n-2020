# Graficando tortas

```
import matplotlib.pyplot as plt
lenguajes = ["Java", "Python", "C#", "JavaScript", "Dart"]
sizes= [10,20,12,18,40]
plt.title ("Población desarrolladores según el lenguaje")
plt.pie (sizes, labels= lenguajes)
plt.show()
```

# Con sombra

```
import matplotlib.pyplot as plt
lenguajes = ["Java", "Python", "C#", "JavaScript", "Dart"]
sizes= [10,20,12,18,40]
plt.title ("Población desarrolladores según el lenguaje")
plt.pie (sizes, labels= lenguajes, shadow=True)
plt.show()
```

# Sombra y sobresaliente

```
import matplotlib.pyplot as plt
lenguajes = ["Java", "Python", "C#", "JavaScript", "Dart"]
sizes= [10,20,12,18,40]
explodeList= [0,0.1,0,0,0]
plt.title ("Población desarrolladores según el lenguaje")
plt.pie (sizes, labels= lenguajes, explode= explodeList, shadow= True)
plt.show()
```

# Edición de ángulo inicial

```
import matplotlib.pyplot as plt
lenguajes = ["Java", "Python", "C#", "JavaScript", "Dart"]
sizes= [10,20,12,18,40]
plt.title ("Población desarrolladores según el lenguaje")
plt.pie (sizes, labels= lenguajes, startangle=45)
plt.savefig ("desarrolladores.png")
plt.show()
```

# Ejercicio práctico

```
import matplotlib.pyplot as plt
preguntaObjetos = "ingrese cuantos elementos caben an la caja : "
preguntaIngresarObjetos = "ingrese nombre objeto : "
preguntaIngresarObjetoOcupacion = "ingrese porcentaje de ocupación del objeto : "
elementos = []
ocupaciones = []
cantidad = int (input(preguntaObjetos))

for i in range (cantidad):
  elemento = input (preguntaIngresarObjetos)
  elementos.append(elemento)
  ocupacion = int (input(preguntaIngresarObjetoOcupacion)) 
  ocupaciones.append (ocupacion)

plt.title ("Población desarrolladores según el lenguaje")
plt.pie (ocupaciones, labels= elementos, shadow=True)
plt.show()
```
